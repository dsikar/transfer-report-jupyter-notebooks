# Load pickle files - nvidia 1

```python
import pickle
import numpy as np
# pickle.dump(steer_err, open( "nvidia1_m120_p120_steer_errlogs_Wed_Nov_25_23_39_22_2020.pickle", "wb" ) )
steer_err = pickle.load( open( "sample_data/nvidia1_m120_p120_steer_err_logs_Fri_Jul_10_09_29_13_2020.pickle", "rb" ) )

# transform 
# steer_err_arr = np.asarray(steer_err)
# steer_err_arr[0][1][0]
#steer_fix = []
#for i in range(0, len(steer_err)):
  #print(steer_err[i][0], steer_err[i][1][0], steer_err[i][1][1], steer_err[i][1][2], steer_err[i][1][3])
#  steer_fix.append([steer_err[i][0], steer_err[i][1][0], steer_err[i][1][1], steer_err[i][1][2], steer_err[i][1][3]])
nvidia1_steer_fix_arr = np.array(steer_err)
```

# Plot

```python
import matplotlib.pyplot as plt
# plot steering errors
SHIFT = 0
MAE = 1
MAPE = 2
MSE = 3
RMSE = 4
# plt.rcParams["figure.figsize"] = (18,3)

plt.plot(nvidia1_steer_fix_arr[:,SHIFT], nvidia1_steer_fix_arr[:,MAE], label='MAE')
plt.plot(nvidia1_steer_fix_arr[:,SHIFT], nvidia1_steer_fix_arr[:,MAPE], label='MAPE')
plt.plot(nvidia1_steer_fix_arr[:,SHIFT], nvidia1_steer_fix_arr[:,MSE], label='MSE')
plt.plot(nvidia1_steer_fix_arr[:,SHIFT], nvidia1_steer_fix_arr[:,RMSE], label='RMSE')

plt.ylabel('nvidia1 Error')
plt.xlabel('RGB Pixel Shift')    
# Set a title of the current axes.

mytitle = 'MAE error logs_Fri_Jul_10_09_29_13_2020'
plt.title(mytitle)
plt.grid(axis='y')
# set limit
#plt.xlim([-5,len(g)+5])
#plt.gca().invert_yaxis()
plt.legend(loc='lower right')
plt.show()
```
